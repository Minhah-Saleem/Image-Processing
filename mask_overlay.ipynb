```python
from skimage import color
from skimage import segmentation
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
path_out = '/home/centos/rabeea/maski/cate2/pred1/cate2-00031.bmp'
path_in= '/home/centos/rabeea/maski/cate2/images/cate2-00031.jpg'
segmentation_results= np.array(Image.open(path_out))
input_image= np.array(Image.open(path_in))
result_image1 = segmentation.mark_boundaries(input_image, segmentation_results, mode='thick')
result_image = color.label2rgb(segmentation_results, result_image1, colors = ('red', 'blue', 'yellow', 'magenta', 'green', 
                   'indigo', 'darkorange', 'cyan', 'pink', 'yellowgreen', 'maroon', 'purple', 'olive', 'teal', 'chocolate',
                    'crimson', 'rosybrown', 'brown', 'salmon', 'coral', 'orangered', 'sienna', 
                    'tan', 'gold', 'wheat', 'lawngreen', 'lightgreen', 'turquoise', 'cadetblue', 'steelblue', 'navy', 
                    'violet', 'blueviolet') , bg_label=0, bg_color=(0, 0, 0), kind='overlay')
# plt.imsave('/home/centos/rabeea/maski/paper/good/cate2-00031.bmp',result_image)
```
